```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df_airport = pd.read_excel("C:\\Users\\Acer\\Desktop\\self-Projects\\05 Airport Data Analysis - Ineuron\\air travel dataset.xlsx")
```

```python
df_airport
```

```python
df_airport.columns
```

```python
df_airport.columns = ['Geometry Coordinates Lat Origin', 'Geometry Coordinates Long Origin',
       'Geometry Coordinates Lat Destination', 'Geometry Coordinates Long Destination', 'Geometry Type',
       'Properties Edtf Cessation', 'Properties Edtf Inception',
       'Properties Flysfo Actual Timestamp', 'Properties Flysfo Airline',
       'Properties Flysfo Base Airline',
       'Properties Flysfo Base Flight Number', 'Date',
       'Properties Flysfo Estimated Timestamp', 'Properties Flysfo Event',
       'Properties Flysfo Flight Number', 'Properties Flysfo Gate', 'Route']
```

```python
df_airport
```

```python
df_airport[['Origin', 'Destination']] = df_airport['Route'].str.split('-', expand=True)
```

```python
df_airport
```

```python
df_airport.drop(['Route'], axis= 1)
```

```python
df_airport.info()
```

```python
df_airport['day_of_week'] = df_airport['Date'].dt.day_name()
```

```python
df_airport
```

```python
df_airport['Distance'] = ((df_airport['Geometry Coordinates Lat Origin'] - df_airport['Geometry Coordinates Lat Destination'])**2 + (df_airport['Geometry Coordinates Long Origin'] - df_airport['Geometry Coordinates Long Destination'])**2)
df_airport['Distance'] = ((df_airport['Distance'])**0.5)*110
```

```python
df_airport.tail()
```

```python
df_airport['City'] = np.where(df_airport['Origin'] != 'SFO', df_airport['Origin'], df_airport['Destination'])
```

```python
df_airport['Origin'].value_counts()
```

```python
df_airport = df_airport.astype({'Date':str})
df_airport['Date'].value_counts().sort_values().plot.bar()
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Date Wise Flight Count',fontsize=25)
plt.xlabel('Date', fontsize=15)
plt.xticks (rotation= 90) 
plt.ylabel('Number of Flights', fontsize=15)
plt.show()
```

```python
df_airport['day_of_week'].value_counts().sort_values().plot.bar()
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Day Wise Flight Count',fontsize=25)
plt.xlabel('Date', fontsize=15)
plt.xticks (rotation= 90) 
plt.ylabel('Number of Flights', fontsize=15)
plt.show()
```

```python
df_airport['Properties Flysfo Gate'].value_counts().sort_values().plot.bar()
plt.rcParams['figure.figsize'] = (24,6)
plt.title('Flight With Gate Count',fontsize=25)
plt.xlabel('Date', fontsize=15)
plt.xticks (rotation= 90) 
plt.ylabel('Number of Flights', fontsize=15)
plt.show()
```

```python
df_airport['Properties Flysfo Event'].value_counts().plot.pie(explode=[0.05,0.1],autopct='%1.1f%%',startangle=180,shadow=True,figsize=(9,9))
plt.title("Departure And Arrival Rate",size = 18)
plt.show()
```

```python
Arrival_data = df_airport[df_airport['Properties Flysfo Event'] == 'A']
Departure_data = df_airport[df_airport['Properties Flysfo Event'] == 'D']
```

```python
df_airport.groupby(['City'])['Distance'].mean().sort_values().tail(10).plot.bar()
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Most Distances From SFO',fontsize=25)
plt.ylabel('Destination', fontsize=15)
plt.xticks (rotation= 0) 
plt.xlabel('Distance', fontsize=15)
plt.show()
```

```python
flight = df_airport.loc[df_airport['Properties Flysfo Base Airline'] != 'UA']
flight ['Properties Flysfo Base Airline'].value_counts().sort_values().tail(10).plot.pie(autopct ='%1.1f%%',shadow = True)
plt.title("Different Base Airline Count",size = 20)
plt.show()
```

```python
flight = df_airport.loc[df_airport['Properties Flysfo Airline'] != 'UA']
flight ['Properties Flysfo Airline'].value_counts().sort_values().tail(10).plot.pie(autopct ='%1.1f%%',shadow = True)
plt.title("Different Airline Count",size = 20)
plt.show()
```

```python
df_airport['delay'] = (df_airport['Properties Flysfo Estimated Timestamp'] - df_airport['Properties Flysfo Actual Timestamp'])/60 #divide by 60 for output in minute
df_airport['delay'].astype({'delay':'int'}) #changing the datatype to int
df1 = df_airport[['delay','City']]
df1.groupby(['City'])['delay'].mean().sort_values().plot.bar()
plt.rcParams['figure.figsize'] = (25,6)
plt.title('Delay of Flights',fontsize=25)
plt.ylabel('Delay in Minutes', fontsize=15) 
plt.xlabel('Destinations', fontsize=15)
plt.show()
```

```python
df1 = df_airport[['delay','Date']]
df1.groupby(['Date'])['delay'].mean().sort_values().plot.bar(x= 'Origin' , y = 'delay')
plt.rcParams['figure.figsize'] = (25,6)
plt.title('Delay of Flights',fontsize=25)
plt.ylabel('Delay in Minutes', fontsize=15) 
plt.xlabel('Date', fontsize=15)
plt.show()
```
